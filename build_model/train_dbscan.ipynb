```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('../../data/processed/Cleaned_AIS_data.csv')
```

```python
df.shape
```

```python
df.columns
```

```python
X = df[['MMSI', 'LAT', 'LON', 'SOG', 'COG', 'Heading', 'VesselType', 
                              'Status', 'TransceiverClass']]
```

```python
# scale the values
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

```python
# demension reduction with PCA
pca = PCA(n_components=2)
pca_X = pca.fit_transform(scaled_X)
```

```python
# plot the data
plt.scatter(x=pca_X[:, 0], y=pca_X[:, 1], marker='.', linewidth=0.1, alpha=0.2)
```

```python
# use dbscan to predict the 
dbscan = DBSCAN()
dbscan.fit_predict(pca_X)
```

The dataset is too large for a DBSCAN model.

Also, according to the [post](https://stackoverflow.com/questions/60713059/dbscan-how-to-cluster-large-dataset-with-one-huge-cluster) in stackoverflow, DBSCAN might not be good for outlier detection
